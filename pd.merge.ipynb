```python
import pandas as pd
df_1 = pd.DataFrame({'ID':[1,2,3,9,8,19,4],
                   'Col_1':[3,4,5,6,7,9,6],
                   'Col_2':[5,6,7,8,5,0,8],
                   'Col_3':['apple','ball','cat','dog','fish','goat','house']})
df_2 = pd.DataFrame({'ID':[1,2,3,9,5,6,4],
                   'Col_A':[3,4,5,6,7,8,5],
                   'Col_B':[5,6,7,8,5,9,78],
                   'Col_3':['apple','ball','cat','dog','elephant','fish','house']})
```

```python
inner_merge = pd.merge(df_1,df_2)
inner_merge
```

```python
pd.merge(df_1,df_2,on='ID')
```

```python
pd.merge(df_1,df_2,on=['ID','Col_3'])
```

```python
pd.merge(df_1,df_2,suffixes=['_l','_r'],left_index=True,right_index=True) #outer no data will be lose joins all rows from both dataframes
```
