```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
nationalities =['Nepal','India','China','Other']
students =[60,20,22,56]
plt.pie(students, labels = nationalities)
plt.show()
```

```python
index = np.arange(4)
plt.bar(index,students,width=0.6,label=nationalities)
```

```python
explode = [0.1,0,0.05,0]
plt.pie(students,labels=nationalities,autopct='%.2f%%',shadow=False,explode=explode)
plt.show()
```

```python
pairs = zip(students,nationalities)
pairs= sorted(list(pairs))
students,nationalities = zip(*pairs)
print(pairs)
```

```python
plt.pie(students,labels=nationalities,autopct='%.2f%%',shadow=False,counterclock= False)
plt.show()
```
