```python
import pandas as pd
import numpy as np
df = pd.read_csv('datasets/cwurData.csv')
df.head()
```

```python
def create_category(ranking):
    if(1<=ranking<=100):
        return "First Tier Top University"
    elif(101<=ranking<=200):
        return "Second Tier Top University"
    elif(201<=ranking<=300):
        return "Third Tier Top University"
    return 'Other Top University'
df['Rank_Level'] = df['world_rank'].apply(lambda x: create_category(x))
df.head()
```

```python
df.pivot_table(values="score",index='country',columns="Rank_Level",aggfunc=[np.mean]).head()
```

```python
df.pivot_table(values="score",index='country',columns="Rank_Level",aggfunc=[np.mean,np.max]).head()
```

```python
df.pivot_table(values="score",index="country",columns="Rank_Level",aggfunc=[np.min,np.max,np.min]).head()
```

```python
df.pivot_table(values='score',index='country',columns='Rank_Level',aggfunc=[np.mean,np.min,np.max],margins=True).head()
```

```python
new_df = df.pivot_table(values='score',index='country',columns='Rank_Level',aggfunc=[np.mean,np.min,np.max],margins=True).head()
```
