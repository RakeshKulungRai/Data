```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-100,100,201)
#x = np.arange(-100,100,1)
y = 0.5 * x ** 2 + 2 *x
y2 = np.sin(x) *2000
y3 = np.log(x) * 1000
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
plt.plot(x, y,'y') #color
plt.show()
```

Sub Plots

```python
x = np.arange(0,100,.2)
y1 = np.sin(x)
y2 = x ** 2 + 2 * x
ax1 = plt.subplot(221) # rows columns indexof_subplots
ax2 = plt.subplot(222)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)


ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y2,'r')
ax4.plot(x,y1,'b')
plt.tight_layout()
plt.show()
```

```python
from matplotlib import style
style.use('grayscale')
x = np.arange(0,30,0.2)
y = np.sin(x)
y2 = np.cos(x)
plt.title('Sin function')
plt.suptitle('Sine Function')
plt.xlabel('number of students')
plt.ylabel('weight of students')
plt.grid(True)
plt.plot(x,y,label='Sine')
plt.plot(x,y2,label ='Cosine',c='y')
plt.legend(loc ='upper left')
plt.show()
```

```python
python = (85, 67, 23, 98)
java = ( 50, 67, 89, 14)
networking = (60, 20, 56, 22)
machine_learning =(58, 23, 40, 87)
people = ['ram','sham','bob','rob']
index = np.arange(4)
plt.bar(index, python,width=0.2,label='python')
plt.bar(index+0.2, java,width=0.2,label='java')
plt.bar(index+0.4, networking,width=0.2,label='networking')
plt.bar(index+0.6, machine_learning,width=0.2,label='machine learing')
plt.title('It skill levels')
plt.xlabel('Person')
plt.ylabel('skill level')
plt.xticks(index,people)
plt.legend(loc='upper left')
plt.ylim(0,120)
plt.show()
```
