```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/census.csv')
df = df[df['SUMLEV']==50]
```

```python
%%timeit -n 3
for state in df['STNAME'].unique():
    avg =np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
   # print('Country in state '+state +'have an average population of'+str(avg))
```

```python
%%timeit -n 3
for group, frame in df.groupby('STNAME'):
    avg =np.average(frame['CENSUS2010POP'])
    print('country in state '+group+'an average population of '+str(avg))
```

```python
df = df.set_index('STNAME')
```

```python
def set_batch_number(item):
    if item[0]<'M':
        return 0
    if item[0]<'Q':
        return 1
for group,frame in df.groupby(set_batch_number):
    print('There are ' +str(len(frame))+ 'records in group'+str(group))
```

```python
df = pd.read_csv('https://bit.ly/3kGDuKx',na_values="?")
```

```python
df.head()
```

```python
#Group by -single group singel column
df.groupby(['pclass'])['survived'].sum() #group by pclass and result is survived
```

```python
#Groupby multiplecolumns
df.groupby(['pclass'])['survived','fare'].sum()
```

```python
df.groupby(['pclass'])['survived','fare'].max()
```

```python
#multiple groups
```

```python
#multiple goups as columns

df.groupby(['pclass','sex'],as_index=False)['survived'].sum()
```

```python
#multiple function
df.groupby(['pclass','sex'],as_index=False)['survived'].agg(['mean','sum','min','max'])
```

```python
df['age_bins']=pd.cut(df['age'],bins =3,labels=('young','middle_age','old'))
```
